```python
import pandas as pd
```

```python
df = pd.read_csv('libraries.csv')
```

```python
# Forward fill the NaN values in the 'Libraries' column
df['Libraries'] = df['Libraries'].fillna(method='ffill')
```

```python
selected_columns = ['Libraries', 'Book Number', 'Books Available']
df_filtered = df[selected_columns]

df_filtered
```

```python
# exclude Framingham, Cary, Millis (too far)

df_filtered = df_filtered[~df_filtered['Libraries'].isin(['Framingham Public Library', 'Cary Memorial Library', 'Millis Public Library', 'Holliston Public Library'])]
df_filtered
```

Finding books unique to each library

```python
book_numbers = df_filtered['Book Number'].value_counts()
```

```python
book_numbers_once = book_numbers[book_numbers == 1].index
```

```python
df_unique_books = df_filtered[df_filtered['Book Number'].isin(book_numbers_once)]
```

```python
df_unique_books
```

```python
df_unique_books.to_csv('unique books.csv')
```
